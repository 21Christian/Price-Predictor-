```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import joblib
from sklearn import tree

# loading the dataset
house_data = pd.read_csv('C:\\Users\\User\\Desktop\\Datasets\\house_sales.csv')
# defining the target data and features
X = house_data.drop(columns=['Price'])
y = house_data['Price']
# training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
# defining the model
model = DecisionTreeClassifier()
# training the model
model.fit(X_train, y_train)
training_predictions = model.predict(X_test)
# using the model to make predictions to guess how much a specific area costs 
model.fit(X.values, y)
predictions = model.predict([[120]])
# visualizing a graph 
tree.export_graphviz(model, out_file = 'house_price_recommender_graph.dot',
#                     feature names so we can set the rules 
                    feature_names=['Price', 'Area'], 
#                     displaying the class for each node
                    class_names=sorted(y.unique()), 
#                     every node has labels to read
                    label ='all',
#                     rounded corners for the boxes
                    rounded=True,
#                     each node is filles with a color
                    filled=True)
predictions 
```
